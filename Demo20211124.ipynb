## 匯率分析

```python
import pandas
dfs = pandas.read_html('https://rate.bot.com.tw/xrt/quote/l6m/USD')
```

```python
df = dfs[0]
df = df.iloc[:,[0,2]]
df.columns = ['date', 'price']
df.set_index('date', inplace=True)
df.sort_index(inplace=True)
df.head()
```

```python
df.plot(kind = 'line', figsize = [20,10])
```

```python
df['ma30'] = df['price'].rolling(window = 30).mean()
```

```python
df.plot(kind = 'line', figsize = [20,10])
```

## Sand Analysis

```python
import requests
res = requests.get('https://www.coingecko.com/price_charts/12129/usd/24_hours.json')
sand = res.json()['stats']
```

```python
import pandas
df = pandas.DataFrame(sand)
```

```python
df.columns =['timestamp', 'price']
```

```python
df.set_index('timestamp',inplace=True)
df.sort_index(inplace=True)
df.head()
```

```python
df.plot(kind = 'line', figsize = [20,10])
```

```python
df['ma24'] = df['price'].rolling(window = 24).mean()
```

```python
df.plot(kind = 'line', figsize = [20,10])
```
